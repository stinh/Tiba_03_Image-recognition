```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=np.full((300, 500, 3), (0,0,0), np.uint8)
cv2.line(m1, (20,10), (20,200), (0,0,255), 2) #起點座標(x,y), 終點座標(x,y) ,顏色(B,G,R) ,粗細

cv2.line(m1, (20,10), (20,200), (0,0,255), 2) 
cv2.rectangle(m1, (40,40), (150,150), (0,255,0), 2)  #左上座標(x,y), 右下座標(x,y) ,顏色(B,G,R) ,粗細
cv2.rectangle(m1, (200,40), (300,150), (0,255,0), -1)

cv2.circle(m1, (100,230), 30, (255,0,0), 2)
cv2.circle(m1, (200,230), 30, (255,0,0), -1)

cv2_imshow(m1)
```

方形

```python
from PIL import ImageFont, ImageDraw, Image
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time # 時間

m1=np.full((1500, 1500, 3), (255,255,255), np.uint8)

for w in range(5,250,15):
  dp.clear_output(True) #關閉前圖，才不會有殘影
  cv2.rectangle(m1, (100+2*w-20,10+3*w),(w,w), (105,w,0), 3)
  cv2_imshow(m1)
  time.sleep(0.3) #設定圖像停留時間，才看得清楚膨脹感
```

每次只有一個圓，清空重劃，不堆疊

```python
from PIL import ImageFont, ImageDraw, Image
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
import time # 時間


for w in range(5,250,15):
  m1=np.full((1500, 1500, 3), (255,255,255), np.uint8) #畫圖的指令放在回圈內，每次就只會產出一個圓，不會堆疊
  dp.clear_output(True) #關閉前圖，才不會有殘影
  cv2.rectangle(m1, (100+2*w-20,10+3*w),(w,w), (105,w,0), 3)
  cv2_imshow(m1)
  time.sleep(0.3) #設定圖像停留時間，才看得清楚膨脹感
```
